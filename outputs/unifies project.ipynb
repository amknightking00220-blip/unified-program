**engagement index**

```python
cols = [
    'JobInvolvement',
    'JobSatisfaction',
    'EnvironmentSatisfaction',
    'RelationshipSatisfaction'
]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[cols] = scaler.fit_transform(df[cols])
```

```python
df['EngagementIndex'] = df[cols].mean(axis=1)
```

```python
df['EngagementIndex'].describe()
```

```python
def engagement_level(score):
    if score < 0.4:
        return 'Low'
    elif score < 0.7:
        return 'Medium'
    else:
        return 'High'

df['EngagementLevel'] = df['EngagementIndex'].apply(engagement_level)
```
